```python
import pandas as pd
from tqdm import tqdm
import json
import re
```

```python
data = []
with open('rotten_tomatoes/output/movie_info.jsonl', 'r') as file:
    for line in file:
        data.append(json.loads(line))

df= pd.DataFrame(data)
```

```python
df[df['director'] == 'Denis Villeneuve']
```

```python
rating_details = []

for r in df['rating']:
    try:
        details = re.search('\([^}]*\)', r).group(0)
        details = details.replace("|", ", ").strip("()")
        rating_details.append(details)
    except:
        rating_details.append(None)

df['rating_details'] = rating_details
df['rating'] = df['rating'].replace('\([^}]*\)', "", regex=True)
```
