```python
import pandas as pd
import json
```

```python
rotten_data = []

with open("output/movies_at_home.txt", 'r') as f_in:
    for line in f_in:
        rotten_data.append(json.loads(line))
```

```python
rotten_df = pd.DataFrame(rotten_data)
```

```python
def add_public_scores(public, df):
    certified = []
    score = []
    scorePercent = []
    sentiment = []

    for scores in rotten_df[f'{public}Score']:
        try:
            certified.append(scores['certified'])
        except KeyError:
            certified.append(None)

        try:
            score.append(scores['score'])
        except KeyError:
            score.append(None)

        try:
            scorePercent.append(scores['scorePercent'])
        except KeyError:
            scorePercent.append(None)

        try:
            sentiment.append(scores['sentiment'])
        except KeyError:
            sentiment.append(None)

    df[f'{public}Certified'] = certified
    df[f'{public}Score'] = score
    df[f'{public}ScorePercent'] = scorePercent
    df[f'{public}Sentiment'] = sentiment

add_public_scores('audience', rotten_df)
add_public_scores('critics', rotten_df)
```
